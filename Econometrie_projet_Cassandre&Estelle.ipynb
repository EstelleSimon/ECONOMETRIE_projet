**Introduction**

Depuis 2020, le secteur de l'énergie a connu une forte inflation qui s’est répercutée sur tous les secteurs de l’économie. Cela s’est accompagné par une hausse des cours boursiers des entreprises majeures du secteur de l’énergie, ce que l’on pourra observer sur le graphique plus bas. Par conséquent les banques centrales, et notamment la Federal Reserve, ont décidé d’augmenter leurs taux d’intérêts directeurs pour permettre de réduire l’inflation. Nous observerons donc l’effet des variations des taux d’intérêts sur les rendements des cours boursiers de l'énergie. Pour cela nous avons choisi le S&P500 Energy car il permet de bien représenter les entreprises du secteur de l’énergie.

Le S&P 500 Energy comprend toutes les sociétés du S&P 500 qui sont classées dans le secteur de l'énergie par le GICS.  La classification GICS (Global Industry Classification Standard) a été créée et est mise à jour par Morgan Stanley (MSCI) et Standard & Poor's (S&P). Il s’agit de l'une des deux principales nomenclatures sectorielles utilisées dans le monde de la finance.

Le secteur de l'énergie est huitième en termes de taille parmi les 11 secteurs du S&P 500.  Au 31 décembre 2018, le secteur représentait 5,32 % du S&P 500. Selon MSCI, “le secteur de l'énergie comprend les entreprises engagées dans l'exploration et la production, le raffinage et la commercialisation, le stockage et le transport du pétrole et du gaz, du charbon et des combustibles consommables. Il comprend également les entreprises qui proposent des équipements et des services dans le domaine du pétrole et du gaz”.

Le S&P500 Energy est composé actuellement de 23 sociétés.  Les deux plus grandes sociétés du secteur étaient Exxon Mobil Corp (XOM) et Chevron Corp (CVX). Les 10 premiers titres du secteur de l'énergie représentent près de 80% de l’indice.

## **VII. ADF**

On effectue le test de Dickey-Fuller augmenté (ADF) en suivant une approche séquentielle pour évaluer la présence de racine unitaire dans les séries temporelles des S&P500 Energy Sector et FedFund. L'objectif est de déterminer si ces séries sont stationnaires ou non et de choisir la méthode de stationnarisation appropriée avant de procéder à l'estimation d'un modèle AR, MA ou ARMA.

Afin d’obtenir un résultat fiable et pour éviter d’omettre une variable dans notre modèle, on commence par estimer le modèle M3 avec constante et tendance déterministe. Puis on estimera le modèle avec une constante et sans tendance (M2) et enfin on estimera le modèle sans tendance déterministe et sans constante (M1).


### ADF FEDFUND ln

```r
library(urca)

# FEDFUND ln M3 (cste + trend)
adf1<-ur.df(y = FEDFUND$log_DataFF,type=c("trend"), lag=15, selectlags = c('AIC'))
summary(adf1)
```

Le critère AIC estime un nombre de retard optimal de 2. Avec les résultats obtenus, on teste la présence d’une racine unitaire. On pose les hypothèses suivantes :

$$\left\lbrace\begin{array}{l}
H_{0}\,:\,\phi_{i}=0\\
H_{1}\,:\,\phi_{i} \neq 0
\end{array} \right.$$

La statistique du test ADF est égale à tADF = -1,254. La statistique de test ADF est supérieure aux seuils de rejets pour les risques de première espèce de 1%, 5% et 10%. On ne peut pas rejeter l’hypothèse nulle de racine unitaire.

Nous testons la significativité de la tendance déterministe. On pose les hypothèses suivantes :

$$\left\lbrace\begin{array}{l} H_{0}\,:\,b=0\\
H_{1}\,:\,b \neq 0
\end{array} \right.$$

Le t-stat de la tendance déterministe est égal à tb= -1,172. Le seuil de rejet qui figure sur la table de Dickey-Fuller est 3.11 pour un risque de première espèce de à 5% et pour un nombre d’observations n = 250. On a donc |tb|=1,172 < 3, 11 : on ne peut pas rejeter l’hypothèse nulle que la tendance déterministe n’est pas significative.

On en déduit que le modèle M3 n’est pas le modèle qui correspond à la série des taux d’intérêts.

##### Causalité de la consommation sur le revenu

$$

### Conclusion

Nous avons étudié les variations du cours de l’indice boursier du S&P500 Energy parallèlement à l’évolution des taux d’intérêts directeurs sur la période de 2000 à aujourd’hui afin d’analyser les liens de corrélations entre les deux variables. Pour étudier les liens entre les deux séries nous avons d’abord estimé un modèle ARMA, qui est un modèle univarié. Pour cela nous avons stationnarité nos série en prenant leurs valeurs en logarithme et en les différenciant une fois. Après cela nous avons estimé un modèle ARMA de la série du S&P 500 Energy en logarithme différencié. Ce modèle nous a permis de reconstituer les prévisions de l’indice. Ainsi nous avons trouvé que l’indice allait fortement augmenter dans les 12 prochains mois. Cependant, cette représentation n'étudiait pas le lien avec les taux d’intérêts. Nous avons donc estimé un modèle VAR qui nous a permis d’étudier ce lien. Nous avons ainsi trouvé qu’il y avait un lien de corrélation significatif d’une variable sur l’autre, c'est-à-dire que le S&P500 Energy a un impact sur les taux d’intérêts directeurs et vice versa. Nous avons vu que les taux d’intérêts directeurs augmentent lorsque l'indice boursier augmente et baisse lorsque que l’indice baisse. Cela est dû à la politique monétaire des banques centrales. Ces dernières ont pour rôle de contenir l’inflation, ainsi lorsqu’il y a de l’inflation elles vont décider d'augmenter les taux afin de contenir l’inflation. A l'inverse, lorsque l’inflation disparaît elles vont revenir à des taux normaux. Les variations de cours du S&P 500 sont un reflet de l’inflation dans le secteur de l’énergie.
D’une autre part on a constaté que la hausse des taux d’intérêts faisait baisser le cours du S&P 500 Energy, mais de manière moindre (puisque le coefficient est de -0,0434 et la fonction d’impulsion réponse illustre bien cela car elle est proche de 0).
Ces impacts sont des impacts de court terme, c’est à dire qu’un choc d’une variable ne va pas impacter durablement l’autre variable. Pour savoir si les variables s’impactent durablement entre elles, nous avons effectué un test de cointégration. Or nous n’avons pas trouvé que l’impact de long terme d’une variable sur l’autre était significatif. On peut donc dire qu’une hausse des taux d’intérêts permettra de faire baisser durant peu de temps le cours boursier du S&P 500 mais pas sur le long terme. D’ailleurs lorsque les banques centrales augmentent les taux d’intérêts, elles le font progressivement sur une longue période de temps (2 ans pendant le covid). Cela permet de faire baisser doucement l’inflation et entraîne ainsi une baisse faible et progressive de l’indice du S&P 500 Energy.
